```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
pythonmarks = [23,20,30,2,10,28,27,1,15]
pythonmarks2 = [30,30,10,21,10,28,24,12,25]
np.random.seed(121)
pythonmarks3 = np.random.randint(1,31,58)
pythonmarks3
```

```python
l=["a","b","c","d","e"]
e=[200,1000,300,400,600]
plt.pie(e,labels=l,shadow=True,autopct="%1.1f%%")
plt.show()
```

```python
l=["a","b","c","d","e"]
e=[200,1000,300,400,600]
ep = [0.0,0.2,0.0,0.0,0.0]
plt.pie(e,labels=l,shadow=True,autopct="%1.1f%%",explode=ep)
plt.show()
```

```python
l=["a","b","c","d","e"]
e=[200,1000,300,400,600]
c=['b','g','r','c','m']
plt.pie(e,colors=c,labels=l,shadow=True,autopct="%1.1f%%",)
plt.show()
```

<h1>bar chart</h1>

```python
x=['bcw','bcx','bcy','bcz']
y=[12,25,34,10]

plt.bar(x,y)
#vertical display bar chart(by default)
```

```python
x=['bcw','bcx','bcy','bcz']
y=[12,25,34,10]

plt.barh(x,y)
#horizontal display bar chart
```

```python
x=['bcw','bcx','bcy','bcz']
y=[12,25,34,10]
l=['1st','2nd','3rd','4th']
plt.bar(x,y)
```

```python
x=['bcw','bcx','bcy','bcz']
y=[12,25,34,10]
c=['b','g','r','c']
l=['1st','2nd','3rd','4th']
plt.bar(x,y,color=c,width=0.5)
```

```python
x=['bcw','bcx','bcy','bcz']

y=[12,25,34,10]
c=['b','g','r','c']
l=['1st','2nd','3rd','4th']
plt.bar(x,y,color=c,width=0.5,linewidth=3,edgecolor="b",linestyle="--")
#alpha: opacity , width: width of bars ,line-width (bar edge) ,
# for label use legend()
```
