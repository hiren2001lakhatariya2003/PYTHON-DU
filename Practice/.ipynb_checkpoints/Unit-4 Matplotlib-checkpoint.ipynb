```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
pythonmarks = [23,20,30,2,10,28,27,1,15]
pythonmarks2 = [30,30,10,21,10,28,24,12,25]
np.random.seed(121)
pythonmarks3 = np.random.randint(1,31,58)
pythonmarks3
```

```python
rollno= [i for i in range(101,159)]
np.random.seed(120)
pythonmarks4 = np.random.randint(1,31,58)
```

```python
plt.plot(rollno,pythonmarks)
plt.show() #open in bottom
```

```python
%matplotlib qt
# open in application 
plt.plot(rollno,pythonmarks)
plt.show() 
```

```python
%matplotlib inline
plt.plot(rollno,pythonmarks)
plt.show() 
# open in bottom 
```

```python
%matplotlib inline
plt.plot(rollno,pythonmarks)
plt.plot(rollno,pythonmarks2)

#multiple line data show
#plt.show() 
plt.savefig('demoplot.pdf',format="pdf")


#when we use show function then do not save the figure in file
#you can only show otherwise save the file
```

```python
%matplotlib inline
plt.plot(rollno,pythonmarks3)
plt.plot(rollno,pythonmarks4)

#multiple line data show
#plt.show() 
plt.savefig('demoplot.pdf',format="pdf")


#when we use show function then do not save the figure in file
#you can only show otherwise save the file
```

```python
%matplotlib qt

a= plt.axes()

a.set_xlim(100,200)
a.set_ylim(0,100)
a.set_xticks(range(101,201,2))
# set the gape of x with print

plt.plot(rollno,pythonmarks3)
plt.plot(rollno,pythonmarks4)

#multiple line data show
#plt.show() 
plt.savefig('demoplot.pdf',format="pdf")


#when we use show function then do not save the figure in file
#you can only show otherwise save the file
```

```python
%matplotlib qt

a= plt.axes()

a.set_xlim(100,200)
a.set_ylim(0,100)
a.set_grid()
# set the gape of x with print

plt.plot(rollno,pythonmarks3)
plt.plot(rollno,pythonmarks4)

#multiple line data show
#plt.show() 
plt.savefig('demoplot.pdf',format="pdf")


#when we use show function then do not save the figure in file
#you can only show otherwise save the file
```
